```python
import numpy as np
from numpy.fft import fft2, ifft2
import matplotlib.pyplot as plt

def mit18336_spectral_ns2d():
    """
    Navier-Stokes equations in vorticity/stream function formulation on the torus
    Version 1.0
    (c) 2008 Jean-Christophe Nave - MIT Department of Mathematics
    jcnave (at) mit (dot) edu

    Dw/Dt = nu.Laplacian(w)
    Laplacian(psi) = -w
    u = psi_y
    v =-psi_x
    """
    nu = 1.0e-3  # viscosity
    NX = 128     # resolution in x
    NY = 128     # resolution in y
    dt = 1e-1    # time step
    TF = 50.0    # final time
    TSCREEN = 25 # screen update interval time
    method = 'crank_nicholson'  # 'Forward_Euler' or 'crank_nicholson'
    initial_condition = 'vortices'  # 'vortices' or 'random'

    dx = 2 * np.pi / NX
    dy = 2 * np.pi / NY
    t = 0.

    # Define initial vorticity distribution
    if initial_condition == 'vortices':
        i, j = np.meshgrid(np.arange(NX), np.arange(NY))
        w = np.exp(-((i*dx-np.pi)**2 + (j*dy-np.pi+np.pi/4)**2)/(0.2)) + \
            np.exp(-((i*dx-np.pi)**2 + (j*dy-np.pi-np.pi/4)**2)/(0.2)) - \
            0.5 * np.exp(-((i*dx-np.pi-np.pi/4)**2 + (j*dy-np.pi-np.pi/4)**2)/(0.4))
    elif initial_condition == 'random':
        w = np.random.uniform(-1, 1, (NX, NY))
    else:
        raise ValueError('Unknown initial conditions!')

    kx, ky = np.meshgrid(np.arange(NX) - np.floor(NX/2), np.arange(NY) - np.floor(NY/2))
    kx *= 2*np.pi/NX
    ky *= 2*np.pi/NY
    kx = 1j * kx
    ky = 1j * ky

    dealias = (np.abs(kx) < 2/3*NX) & (np.abs(ky) < 2/3*NY)  # 2/3 rule dealiasing

    ksquare_viscous = kx**2 + ky**2        # Laplacian in Fourier space
    ksquare_poisson = ksquare_viscous.copy()
    ksquare_poisson[0, 0] = 1e-15          # Set a small positive value to avoid division by zero
    w_hat = fft2(w)

    cutoff = 16  # Set a maximum wavenumber cutoff to restrict the basis
    w_hat[(np.abs(kx) > cutoff) | (np.abs(ky) > cutoff)] = 0

    k = 0
    while t < TF:
        k += 1

        # Compute the stream function and get the velocity and gradient of vorticity
        psi_hat = -w_hat / ksquare_poisson  # Solve Poisson's Equation
        u = np.real(ifft2(ky * psi_hat))    # Compute y derivative of stream function
        v = np.real(ifft2(-kx * psi_hat))   # Compute -x derivative of stream function
        w_x = np.real(ifft2(kx * w_hat))    # Compute x derivative of vorticity
        w_y = np.real(ifft2(ky * w_hat))    # Compute y derivative of vorticity

        conv = u * w_x + v * w_y  # evaluate the convective derivative (u,v).grad(w)
        conv_hat = fft2(conv)     # go back to Fourier space

        conv_hat *= dealias  # Perform spherical dealiasing 2/3 rule

        # Compute Solution at the next step
        if method == 'forward_euler':
            w_hat_new = w_hat + dt * (nu * ksquare_viscous * w_hat - conv_hat)
        elif method == 'crank_nicholson':
            w_hat_new = ((1/dt + 0.5*nu*ksquare_viscous)/(1/dt - 0.5*nu*ksquare_viscous))*w_hat - \
                        (1./(1/dt - 0.5*nu*ksquare_viscous))*conv_hat
        else:
            raise ValueError('Unknown method!')

        t += dt

        if k == TSCREEN:
            # Go back in real space omega in real space for plotting
            w = np.real(ifft2(w_hat_new))
            plt.contourf(w, 50)
            plt.colorbar()
            plt.shading('flat')
            plt.colormap('jet')
            plt.title(f"{t:.2f}")
            plt.show()
            k = 0

        w_hat = w_hat_new

mit18336_spectral_ns2d()
```
