```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from tqdm import tqdm
import numpy as np
# Helper functions
def pair(t):
    return t if isinstance(t, tuple) else (t, t)

# Define Patch Embedding
class PatchEmbedding(nn.Module):
    def __init__(self, in_channels, patch_size, emb_dim, img_size):
        super().__init__()
        patch_size = pair(patch_size)
        self.patch_size = patch_size
        self.num_patches = (img_size[0] // patch_size[0]) * (img_size[1] // patch_size[1])
        self.patch_dim = in_channels * patch_size[0] * patch_size[1]
        self.patch_embedding = nn.Linear(self.patch_dim, emb_dim)

    def forward(self, x):
        B, C, H, W = x.shape
        x = x.view(B, C, H // self.patch_size[0], self.patch_size[0], W // self.patch_size[1], self.patch_size[1])
        x = x.permute(0, 2, 4, 1, 3, 5).reshape(B, -1, self.patch_dim)
        x = self.patch_embedding(x)
        return x

# Define Transformer Encoder Layer
class TransformerEncoderLayer(nn.Module):
    def __init__(self, emb_dim, num_heads, mlp_dim, dropout=0.1):
        super().__init__()
        self.layer_norm1 = nn.LayerNorm(emb_dim)
        self.attn = nn.MultiheadAttention(emb_dim, num_heads, dropout=dropout)
        self.layer_norm2 = nn.LayerNorm(emb_dim)
        self.mlp = nn.Sequential(
            nn.Linear(emb_dim, mlp_dim),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(mlp_dim, emb_dim),
            nn.Dropout(dropout)
        )

    def forward(self, x):
        attn_out, _ = self.attn(x, x, x)
        x = x + attn_out
        x = self.layer_norm1(x)
        mlp_out = self.mlp(x)
        x = x + mlp_out
        x = self.layer_norm2(x)
        return x

# Define the Vision Transformer
class VisionTransformer(nn.Module):
    def __init__(self, in_channels=1, img_size=(192, 192), patch_size=16, emb_dim=768, depth=12, num_heads=8, mlp_dim=3072, out_dim=(384, 384)):
        super().__init__()
        self.patch_embedding = PatchEmbedding(in_channels, patch_size, emb_dim, img_size)
        self.cls_token = nn.Parameter(torch.randn(1, 1, emb_dim))
        self.position_embedding = nn.Parameter(torch.randn(1, self.patch_embedding.num_patches + 1, emb_dim))
        self.transformer_encoder = nn.ModuleList([TransformerEncoderLayer(emb_dim, num_heads, mlp_dim) for _ in range(depth)])
        self.layer_norm = nn.LayerNorm(emb_dim)
        
        # Upsampling layers to match output dimension
        self.output_layer = nn.Sequential(
            nn.Linear(emb_dim, out_dim[0] * out_dim[1]),
            nn.GELU(),
            nn.Unflatten(1, out_dim)
        )

    def forward(self, x):
        B = x.size(0)
        x = self.patch_embedding(x)
        cls_tokens = self.cls_token.expand(B, -1, -1)
        x = torch.cat((cls_tokens, x), dim=1)
        x = x + self.position_embedding
        
        # Transformer Encoder
        for layer in self.transformer_encoder:
            x = layer(x)
        
        x = self.layer_norm(x[:, 0])  # Take the [CLS] token
        output = self.output_layer(x)
        return output
```

```python
model = VisionTransformer(in_channels=1, img_size=(192, 192), patch_size=16, emb_dim=768, depth=12, num_heads=8, mlp_dim=3072, out_dim=(384, 384))
```

```python
input_images = np.load('/pscratch/sd/h/hbassi/2d_vlasov_data_coarse_scale_64.npy') 
input_images = torch.Tensor(input_images)
target_images = np.load('/pscratch/sd/h/hbassi/2d_vlasov_data_fine_scale_128.npy') 
target_images = torch.Tensor(target_images).squeeze(1)
print(target_images.shape)
dataset = TensorDataset(input_images, target_images)
dataloader = DataLoader(dataset, batch_size=8)
```

```python
output = model(input_images)
print("Output shape:", output.shape)  # Should be (78, 1, 384, 384)
```

```python
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)
```

```python
# Training loop
def train(model, dataloader, criterion, optimizer, num_epochs=10):
    model.train()
    for epoch in range(num_epochs):
        running_loss = 0.0
        for images, targets in tqdm(dataloader, desc=f"Epoch {epoch+1}/{num_epochs}"):
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(dataloader)}")
```

```python
train(model, dataloader, criterion, optimizer, num_epochs=10)
```
