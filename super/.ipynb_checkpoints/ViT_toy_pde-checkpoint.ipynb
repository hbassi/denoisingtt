```python
import itertools as it
import pickle
import numpy as np
import matplotlib.pyplot as plt
import torch 
import torch.nn as nn
import torch.nn.functional as F
import math
from torch.utils.data import DataLoader, TensorDataset, random_split
#import torch.optim as optim
#import utilities as ut
import attention as att
import torch.optim as optim
from tqdm import trange
```

```python
# For same initialization of weights/biases of model
torch.manual_seed(seed=999)
torch.set_float32_matmul_precision('high')
```

```python
#input_CG = np.load('/pscratch/sd/h/hbassi/toy_pde_data_fine_scale_blurred.npy')
input_CG = np.load('/pscratch/sd/h/hbassi/toy_pde_data_coarse_scale.npy')
target_FG = np.load('/pscratch/sd/h/hbassi/toy_pde_data_fine_scale.npy')
input_CG = np.expand_dims(input_CG, 1)
target_FG = np.expand_dims(target_FG, 1)

input_CG.shape, target_FG.shape
# input_CG_torch, target_CG_torch = att.data_batch_to_torch(input_CG, target_CG)
# print(len(whole_S_data))
# print('CGed input shape is', input_CG_torch[-1].shape, "\n Number of U is:",input_CG_torch[-1].shape[0],
#      "\n Number of Nk is:",input_CG_torch[-1].shape[1], "\n Number of Ntau is:",input_CG_torch[-1].shape[2])
# print('CGed target shape is', target_CG_torch[-1].shape)
```

```python
### Define parameters in the NN 
# previous was 256 and 64
bond_dim = 256
num_heads = 64  ## bond_dim/num_heads must be a integer
num_layers = 2
#num_Nk = len(whole_S_data)
mask = None  # Optional: provide a mask if needed
input_dim = input_CG.shape[2]
output_dim = target_FG.shape[2]
print(input_dim, output_dim) 
```

```python
vit_model_regression = att.VisionTransformer(
    img_height=input_dim,        # Image height
    img_width=input_dim,       # Image width (non-square image)
    patch_size_h=5,      # Patch size for height
    patch_size_w=5,     # Patch size for width
    input_dim=1,         # 1 for single channel images
    num_heads=num_heads,         
    num_layers=num_layers,        
    bond_dim=bond_dim,        
    output_dim=output_dim**2,
    dropout = 0.1
)
```

```python
device = 'cuda'
input_CG_tensor = torch.tensor(input_CG, dtype=torch.float32).to(device)
target_FG_tensor = torch.tensor(target_FG, dtype=torch.float32).to(device)
```

```python
# Create the full dataset
dataset = TensorDataset(input_CG_tensor[:4000], target_FG_tensor[:4000])

# 80% training, 20% validation
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size

# Split the dataset
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

# Create DataLoaders for training and validation
train_dataloader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=64, shuffle=False)
```

```python
# Define loss function and optimizer
criterion = nn.L1Loss()
optimizer = optim.Adam(vit_model_regression.parameters(), lr=0.001)
```

```python
# Training loop
num_epochs = 2500
```

```python
vit_model_regression = vit_model_regression.to(device)
vit_model_regression = torch.compile(vit_model_regression)
```

```python
for epoch in trange(num_epochs):
    vit_model_regression.train()
    running_loss = 0.0

    for batch_idx, (data, target) in enumerate(train_dataloader):
        output = vit_model_regression(data)
        target_flat = target.view(-1, 100 * 100)
        output = output.view(-1, 100* 100)
        #print(target_flat.shape, output.shape)
        loss = criterion(output, target_flat)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
    if epoch % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss / len(train_dataloader):.8f}')

        # Validation phase
        vit_model_regression.eval()
        val_loss = 0.0
        with torch.no_grad():
            for data, target in val_dataloader:
                output = vit_model_regression(data)
                target_flat = target.view(-1, 100 * 100)
                output = output.view(-1, 100* 100)
                loss = criterion(output, target_flat)
                val_loss += loss.item()

        print(f'Validation Loss: {val_loss / len(val_dataloader):.8f}')
        print('=============================================================================')
```

```python
# Load data for test result
input_CG = np.load('/pscratch/sd/h/hbassi/toy_pde_data_coarse_scale.npy')
#input_CG = np.load('/pscratch/sd/h/hbassi/toy_pde_data_fine_scale_blurred.npy')
target_FG = np.load('/pscratch/sd/h/hbassi/toy_pde_data_fine_scale.npy')
```

```python
input_CG_test = input_CG[4000:]  
target_FG_test = target_FG[4000:]  
```

```python
input_CG_test = np.expand_dims(input_CG_test, 1)  
target_FG_test = np.expand_dims(target_FG_test, 1)  
```

```python
input_CG_test_tensor = torch.tensor(input_CG_test, dtype=torch.float32)
target_FG_test_tensor = torch.tensor(target_FG_test, dtype=torch.float32)
```

```python
vit_model_regression.eval().cpu()
```

```python
predictions = []
targets = []
inputs = []  

with torch.no_grad():
 
    batch_size = 64
    for i in range(0, input_CG_test_tensor.shape[0], batch_size):
        data = input_CG_test_tensor[i:i + batch_size]
        target = target_FG_test_tensor[i:i + batch_size]

       
        #print(f"Batch Data shape: {data.shape}, Target shape: {target.shape}")

        output = vit_model_regression(data)  
     
        #print(f"Output shape before reshaping: {output.shape}")

        # Reshape the output back to (batch_size, 1, 100, 100)
        output = output.view(output.shape[0], 1, 100, 100)  

       
        predictions.append(output.numpy())
        targets.append(target.numpy())
        inputs.append(data.numpy())  
 
```

```python
# Check if predictions are collected
if predictions:
    # Convert predictions, targets, and inputs to numpy arrays
    predictions = np.concatenate(predictions, axis=0)  # Shape: (100, 1, 100, 100)
    targets = np.concatenate(targets, axis=0)  # Shape: (100, 1, 100, 100)
    inputs = np.concatenate(inputs, axis=0)  # Shape: (100, 1, 50, 50)

    # Check final shapes
    #print("Final Predictions shape:", predictions.shape)
    #print("Final Targets shape:", targets.shape)
    #print("Final Inputs shape:", inputs.shape)

    # Visualize a few predictions vs targets and inputs
    num_samples = 4  # Number of samples to visualize
    indices = np.random.choice(len(predictions), num_samples, replace=False)

    plt.figure(figsize=(15, 12))  # Adjust height for the extra row
    for i, idx in enumerate(indices):
        # Plot the input
        plt.subplot(4, num_samples, i + 1)
        plt.imshow(inputs[idx, 0], cmap='hot')  # Use the first channel
        plt.title('Input')
        plt.axis('off')
        
        # Plot the target
        plt.subplot(4, num_samples, i + 1 + num_samples)
        plt.imshow(targets[idx, 0], cmap='hot')  # Use the first channel
        plt.title('Target')
        plt.axis('off')
        
        # Plot the prediction
        plt.subplot(4, num_samples, i + 1 + 2 * num_samples)
        plt.imshow(predictions[idx, 0], cmap='hot')  # Use the first channel
        plt.title('Prediction')
        plt.axis('off')

        # Calculate the error and plot the error heatmap with color bar
        error = np.abs(predictions[idx, 0] - targets[idx, 0])
        ax = plt.subplot(4, num_samples, i + 1 + 3 * num_samples)
        im = ax.imshow(error, cmap='viridis')  # Viridis colormap for error heatmap
        plt.title('Error Heatmap')
        plt.axis('off')
        
        # Add color bar only to the error plot
        plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)  # Adjust color bar size and position

    plt.tight_layout()
    #plt.savefig('vit_GS_model_test_result_blurred.png')
    plt.show()
else:
    print("No predictions were generated. Please check the model and data.")
```
