```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import trange
```

```python
# Parameters
L = 100.0          # Length of the domain
Nx = 100           # Number of grid points in x
Ny = 100           # Number of grid points in y
Nt = 1500          # Number of time steps
D_u = 0.16         # Diffusion coefficient for U
D_v = 0.08         # Diffusion coefficient for V
# cary the feed rate between 0.029 and 0.045
F = 0.045          # Feed rate of U
k = 0.06       # Kill rate of V
dt = 1.0           # Time step
dx = L / (Nx - 1)  # Spatial step (same for x and u)
```

```python
# Low-rank approximation function using SVD
def low_rank_approximation(matrix, rank):
    U, S, VT = np.linalg.svd(matrix, full_matrices=False)
    S = np.diag(S[:rank])
    return U[:, :rank] @ S @ VT[:rank, :]
```

```python
# Laplacian operator for 2D
def compute_laplacian(Z, dx):
    laplacian = (np.roll(Z, 1, axis=0) + np.roll(Z, -1, axis=0) +
                 np.roll(Z, 1, axis=1) + np.roll(Z, -1, axis=1) - 4 * Z) / dx**2
    return laplacian
```

```python
# Initialize U and V
U = np.ones((Nx, Ny))
V = np.zeros((Nx, Ny))
start, end = 40, 60
U[start:end, start:end] = 0.4
V[start:end, start:end] = 0.25
# Arrays to store dynamics
dynamics_U = np.zeros((Nt, Nx, Ny))
dynamics_V = np.zeros((Nt, Nx, Ny))
# Low-rank solver
rank = 32  # Desired rank for approximation
for n in trange(Nt):
    # Compute laplacians
    laplacian_U = compute_laplacian(U, dx)
    laplacian_V = compute_laplacian(V, dx)

    # Update U and V
    U_new = U + (D_u * laplacian_U - U * V**2 + F * (1 - U)) * dt
    V_new = V + (D_v * laplacian_V + U * V**2 - (F + k) * V) * dt

    # Apply low-rank approximation to U and V
    U = low_rank_approximation(U_new, rank)
    V = low_rank_approximation(V_new, rank)
    # Store dynamics
    dynamics_U[n] = U
    dynamics_V[n] = V
# Visualization of selected timesteps
time_steps_to_visualize = [0, Nt // 4, Nt // 2, 3 * Nt // 4, Nt - 1]  # Example timesteps
fig, axes = plt.subplots(len(time_steps_to_visualize), 2, figsize=(12, len(time_steps_to_visualize) * 4))

for i, t in enumerate(time_steps_to_visualize):
    im_U = axes[i, 0].imshow(dynamics_U[t], extent=[0, L, 0, L], cmap="viridis")
    axes[i, 0].set_title(f"U at timestep {t}")
    plt.colorbar(im_U, ax=axes[i, 0])  # Add colorbar for U
    
    im_V = axes[i, 1].imshow(dynamics_V[t], extent=[0, L, 0, L], cmap="plasma")
    axes[i, 1].set_title(f"V at timestep {t}")
    plt.colorbar(im_V, ax=axes[i, 1])  # Add colorbar for V

plt.tight_layout()
plt.savefig('./figures/example_GS_dynamics_TT_compression_bd=32.png')
plt.show()

```
